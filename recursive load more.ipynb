```python
from bs4 import BeautifulSoup; import requests; import pandas as pd

pd.set_option('display.max_colwidth', None); pd.set_option('display.max_rows', None)
df1 = pd.DataFrame(columns=['url','name'])

def f(url, df):
    s = BeautifulSoup(requests.get(url).text, 'html.parser')
    t = s.find_all(class_='name')
    for i in range(len(t)):
        f.u = df = df.append({'url': t[i].a['href'], 'name': t[i].text.replace('<p>', '')}, ignore_index=True)
    if s.find_all(class_='load-more') != []: f(s.find_all(class_='load-more')[0]['href'], df) 

f('https://www.fbi.gov/wanted/fugitives', df1)

print(f.u['url']) 
print(f.u['name'])
```
